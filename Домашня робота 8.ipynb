### 1.1 –ó–∞–≤–¥–∞–Ω–Ω—è 1.1

–†–æ–∑–≥–ª—è–Ω–µ–º–æ –¥–≤—ñ –Ω–µ–∑–∞–ª–µ–∂–Ω—ñ –≤–∏–ø–∞–¥–∫–æ–≤—ñ –≤–µ–ª–∏—á–∏–Ω–∏ $\xi_1, \xi_2$ –∑—ñ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –µ–∫—Å–ø–æ–Ω–µ–Ω—Ü—ñ–π–Ω–æ–≥–æ —Ä–æ–∑–ø–æ–¥—ñ–ª—É.
–¢–æ–±—Ç–æ $\xi_1, \xi_2 \sim Exp(1)$.

–Ø–∫–∏–π —Ä–æ–∑–ø–æ–¥—ñ–ª –º–∞—î –≤–∏–ø–∞–¥–∫–æ–≤–∞ –≤–µ–ª–∏—á–∏–Ω–∞ $\displaystyle \eta = \frac{\xi_1}{\xi_1 + \xi_2}$?
1. –ó–≥–µ–Ω–µ—Ä—É–π—Ç–µ $n = 10000$ –∑–Ω–∞—á–µ–Ω—å –≤–∏–ø–∞–¥–∫–æ–≤–∏—Ö –≤–µ–ª–∏—á–∏–Ω $\xi_1, \xi_2$;
2. –û—Ç—Ä–∏–º–∞–π—Ç–µ $n = 10000$ –∑–Ω–∞—á–µ–Ω—å –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ $\eta$;
3. –ó–æ–±—Ä–∞–∑—ñ—Ç—å –Ω–∞ –≥—Ä–∞—Ñ—ñ–∫—É:
   - –Ω–æ—Ä–º–æ–≤–∞–Ω—É –≥—ñ—Å—Ç–æ–≥—Ä–∞–º—É –æ—Ç—Ä–∏–º–∞–Ω–∏—Ö $n = 10000$ –∑–Ω–∞—á–µ–Ω—å –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ $\eta$;
   - —Ñ—É–Ω–∫—Ü—ñ—é —â—ñ–ª—å–Ω–æ—Å—Ç—ñ –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ $\eta$.

**–ó–∞—É–≤–∞–∂–µ–Ω–Ω—è.** –ü–µ—Ä–µ–¥–±–∞—á–∞—î—Ç—å—Å—è –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è —Ç–µ–æ—Ä–µ–º–∏ –ø—Ä–æ –æ–±—Ä–∞—Ö—É–Ω–æ–∫ —â—ñ–ª—å–Ω–æ—Å—Ç—ñ —Ñ—É–Ω–∫—Ü—ñ—ó –≤—ñ–¥ –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ (—á–µ—Ä–µ–∑ —è–∫–æ–±—ñ–∞–Ω –æ–±–µ—Ä–Ω–µ–Ω–æ–≥–æ –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è).

$$f^x(x1, x2)=e^{-x1}*e^{-x2}I_{x1\geq0}I_{x2\geq0}$$$
$$$y1=x1$$$
$$$y2=\frac{x1}{x1 + x2}$$$
$$$f^{y}(y1, y2)=e^{-y1}*e^{y1(1-\frac{1}{y2})}I_{y1\geq0}I_{0\leq y2\leq1}\frac{-y1}{y2^2}
$$$f^{y2}(y2)=\int_{0}^{\infty}e^{-y1}*e^{y1(1-\frac{1}{y2})}I_{0\leq y2\leq1}dy1=1I_{0\leq y2\leq1}$$

```python
# –ú—ñ—Å—Ü–µ –¥–ª—è –í–∞—à–æ–≥–æ –∫–æ–¥—É

from scipy.stats import expon, uniform
xi1=expon()
xi2=expon()
rvs1=xi1.rvs(10000)
rvs2=xi2.rvs(10000)
x=np.arange(-1, 5+0.01, 0.01)
y=xi1.pdf(x)
fig, axs=plt.subplots(1, 2, figsize=(20, 6))
axs[0].hist(rvs1, density=True, bins='auto')#axs[0].hist(rvs, dencity=True, bins='auto')
axs[0].plot(x, y, 'r')
eta=rvs1/(rvs1+rvs2)
axs[1].hist(eta, density=True, bins='auto')#axs[0].hist(rvs, dencity=True, bins='auto')
axs[1].plot(x, y, 'r')
axs[1].plot(x, 1*(x>=0)*(x<=1), 'g')
plt.show()
```

–¢–æ–±—Ç–æ $\xi_1, \xi_2 \sim Exp(1)$.

2. –û—Ç—Ä–∏–º–∞–π—Ç–µ $n = 10000$ –∑–Ω–∞—á–µ–Ω—å –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ $\eta$;
3. –ó–æ–±—Ä–∞–∑—ñ—Ç—å –Ω–∞ –≥—Ä–∞—Ñ—ñ–∫—É:

**–ó–∞—É–≤–∞–∂–µ–Ω–Ω—è.** –ü–µ—Ä–µ–¥–±–∞—á–∞—î—Ç—å—Å—è –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è —Ç–µ–æ—Ä–µ–º–∏ –ø—Ä–æ –æ–±—Ä–∞—Ö—É–Ω–æ–∫ —â—ñ–ª—å–Ω–æ—Å—Ç—ñ —Å—É–º–∏ –¥–≤–æ—Ö –Ω–µ–∑–∞–ª–µ–∂–Ω–∏—Ö –≤–∏–ø–∞–¥–∫–æ–≤–∏—Ö –≤–µ–ª–∏—á–∏–Ω (—á–µ—Ä–µ–∑ –∑–≥–æ—Ä—Ç–∫—É —â—ñ–ª—å–Ω–æ—Å—Ç–µ–π).

$$ùëìùúÇ(ùë•)=ùëì\xi1+\xi2(x)=\int_{0}^{x}ùëí^{-ùë•+t}*ùëí^{‚àít}dt=xùëí^{‚àíx}, x\geq0$$


```python
# –ú—ñ—Å—Ü–µ –¥–ª—è –í–∞—à–æ–≥–æ –∫–æ–¥—É

from scipy.stats import expon, uniform
xi1=expon()
xi2=expon()
rvs1=xi1.rvs(10000)
rvs2=xi2.rvs(10000)
x=np.arange(-1, 5+0.01, 0.01)
y=xi1.pdf(x)
fig, axs=plt.subplots(1, 2, figsize=(20, 6))
axs[0].hist(rvs1, density=True, bins='auto')#axs[0].hist(rvs, dencity=True, bins='auto')
axs[0].plot(x, y, 'r')
eta=(rvs1+rvs2)
axs[1].hist(eta, density=True, bins='auto')#axs[0].hist(rvs, dencity=True, bins='auto')
axs[1].plot(x, y, 'r')
eta_pdf=x*np.exp(-x)*(x>=0)
axs[1].plot(x, eta_pdf, 'g')
plt.show()
```

### 2.1 –ó–∞–≤–¥–∞–Ω–Ω—è 2.1

–†–æ–∑–≥–ª—è–Ω–µ–º–æ —Ç—Ä–∏ –Ω–µ–∑–∞–ª–µ–∂–Ω—ñ –≤–∏–ø–∞–¥–∫–æ–≤—ñ –≤–µ–ª–∏—á–∏–Ω–∏ $\xi_1, \xi_2, \theta$ —Ç–∞–∫—ñ,
—â–æ $\xi_1, \xi_2 \sim N(0, 1), \theta \sim Uniform(0, 2 \pi)$.

–Ø–∫–∏–π —Ä–æ–∑–ø–æ–¥—ñ–ª –º–∞—î –≤–∏–ø–∞–¥–∫–æ–≤–∞ –≤–µ–ª–∏—á–∏–Ω–∞ $\displaystyle \zeta = \xi_1 \cos \theta + \xi_2 \sin \theta$?
1. –ó–≥–µ–Ω–µ—Ä—É–π—Ç–µ $n = 10000$ –∑–Ω–∞—á–µ–Ω—å –≤–∏–ø–∞–¥–∫–æ–≤–∏—Ö –≤–µ–ª–∏—á–∏–Ω $\xi_1, \xi_2, \theta$;
2. –û—Ç—Ä–∏–º–∞–π—Ç–µ $n = 10000$ –∑–Ω–∞—á–µ–Ω—å –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ $\zeta$;
3. –ó–æ–±—Ä–∞–∑—ñ—Ç—å –Ω–∞ –≥—Ä–∞—Ñ—ñ–∫—É:
   - –Ω–æ—Ä–º–æ–≤–∞–Ω—É –≥—ñ—Å—Ç–æ–≥—Ä–∞–º—É –æ—Ç—Ä–∏–º–∞–Ω–∏—Ö $n = 10000$ –∑–Ω–∞—á–µ–Ω—å –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ $\zeta$;
   - —Ñ—É–Ω–∫—Ü—ñ—é —â—ñ–ª—å–Ω–æ—Å—Ç—ñ –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ $\zeta$.

**–ó–∞—É–≤–∞–∂–µ–Ω–Ω—è.** –ü–µ—Ä–µ–¥–±–∞—á–∞—î—Ç—å—Å—è –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è —Ç–µ–æ—Ä–µ–º–∏ –ø—Ä–æ –ª—ñ–Ω—ñ–π–Ω–µ –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –Ω–æ—Ä–º–∞–ª—å–Ω–∏—Ö –≤–µ–ª–∏—á–∏–Ω.

$\zeta$  - —Ç–µ–∂ –Ω–æ—Ä–º–∞–ª—å–Ω–∏–π —Ä–æ–∑–ø–æ–¥—ñ–ª

```python
# –ú—ñ—Å—Ü–µ –¥–ª—è –í–∞—à–æ–≥–æ –∫–æ–¥—É
from scipy.stats import norm, uniform
xi1=norm()
xi2=norm()
zeta=norm()
theta=uniform(loc=0, scale=2*np.pi)
xi1_rvs=xi1.rvs(10000)
xi2_rvs=xi2.rvs(10000)
theta_rvs=theta.rvs(10000)
zeta_rvs=xi1_rvs*np.cos(theta_rvs)+xi2_rvs*np.sin(theta_rvs)
x=np.arange(-4, 5+0.01, 0.01)
y=zeta.pdf(x)
fig, axs=plt.subplots(1, 1, figsize=(20, 6))
axs.hist(zeta_rvs, density=True, bins='auto')#axs[0].hist(rvs, dencity=True, bins='auto')
axs.plot(x, y, 'r')



plt.show()
```

### 2.2 –ó–∞–≤–¥–∞–Ω–Ω—è 2.2

–†–æ–∑–≥–ª—è–Ω–µ–º–æ –¥–≤—ñ –Ω–µ–∑–∞–ª–µ–∂–Ω—ñ –≤–∏–ø–∞–¥–∫–æ–≤—ñ –≤–µ–ª–∏—á–∏–Ω–∏ $\xi_1, \xi_2$ –∑—ñ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–æ–∑–ø–æ–¥—ñ–ª—É.
–¢–æ–±—Ç–æ $\xi_1, \xi_2 \sim N(0, 1)$.

1. –ó–≥–µ–Ω–µ—Ä—É–π—Ç–µ $n = 10000$ –∑–Ω–∞—á–µ–Ω—å –≤–∏–ø–∞–¥–∫–æ–≤–∏—Ö –≤–µ–ª–∏—á–∏–Ω $\xi_1, \xi_2$;
2. –û—Ç—Ä–∏–º–∞–π—Ç–µ $n = 10000$ –∑–Ω–∞—á–µ–Ω—å –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ $\zeta$;

4. –ó–æ–±—Ä–∞–∑—ñ—Ç—å –Ω–∞ –≥—Ä–∞—Ñ—ñ–∫—É:
   - –Ω–æ—Ä–º–æ–≤–∞–Ω—É –≥—ñ—Å—Ç–æ–≥—Ä–∞–º—É –æ—Ç—Ä–∏–º–∞–Ω–∏—Ö $n = 10000$ –∑–Ω–∞—á–µ–Ω—å –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ $\zeta$;
   - —Ñ—É–Ω–∫—Ü—ñ—é —â—ñ–ª—å–Ω–æ—Å—Ç—ñ –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ $\zeta$;
5. –ó–æ–±—Ä–∞–∑—ñ—Ç—å –Ω–∞ –≥—Ä–∞—Ñ—ñ–∫—É:
   - –Ω–æ—Ä–º–æ–≤–∞–Ω—É –≥—ñ—Å—Ç–æ–≥—Ä–∞–º—É –æ—Ç—Ä–∏–º–∞–Ω–∏—Ö $n = 10000$ –∑–Ω–∞—á–µ–Ω—å –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ $\eta$;
   - —Ñ—É–Ω–∫—Ü—ñ—é —â—ñ–ª—å–Ω–æ—Å—Ç—ñ –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ $\eta$.

**–ó–∞—É–≤–∞–∂–µ–Ω–Ω—è 1.** –ü–µ—Ä–µ–¥–±–∞—á–∞—î—Ç—å—Å—è –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è —Ç–µ–æ—Ä–µ–º–∏ –ø—Ä–æ –æ–±—Ä–∞—Ö—É–Ω–æ–∫ —â—ñ–ª—å–Ω–æ—Å—Ç—ñ —Ñ—É–Ω–∫—Ü—ñ—ó –≤—ñ–¥ –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ (—á–µ—Ä–µ–∑ —è–∫–æ–±—ñ–∞–Ω –æ–±–µ—Ä–Ω–µ–Ω–æ–≥–æ –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è).

**–ó–∞—É–≤–∞–∂–µ–Ω–Ω—è 2.** –ù–∞—Å–ø—Ä–∞–≤–¥—ñ, —Ä–æ–∑–ø–æ–¥—ñ–ª –æ—Ç—Ä–∏–º–∞–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω—å –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ $\eta$ –±—É–¥–µ —Ç—Ä–æ—Ö–∏ –≤—ñ–¥—Ä—ñ–∑–Ω—è—Ç–∏—Å—è –≤—ñ–¥ –ø–æ—Ä–∞—Ö–æ–≤–∞–Ω–æ–≥–æ —Ç–µ–æ—Ä–µ—Ç–∏—á–Ω–æ (–±—É–¥—É—Ç—å —ñ–Ω—à—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏).
–¢–∞–∫ –≤–∏—Ö–æ–¥–∏—Ç—å –∑–∞ —Ä–∞—Ö—É–Ω–æ–∫ —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—ó —Ñ—É–Ω–∫—Ü—ñ—ó $\arccos$ (—ó—ó –æ–±–ª–∞—Å—Ç—ñ –∑–Ω–∞—á–µ–Ω—å).

```python
# –ú—ñ—Å—Ü–µ –¥–ª—è –í–∞—à–æ–≥–æ –∫–æ–¥—É

from scipy.stats import norm, uniform
xi1=norm()
xi2=norm()
xi1_rvs=xi1.rvs(10000)
xi2_rvs=xi2.rvs(10000)

zeta_rvs=np.sqrt(xi1_rvs**2+xi2_rvs**2)
#print(xi1_rvs/zeta_rvs)
eta_rvs=np.arccos(xi1_rvs/zeta_rvs)
fig, axs=plt.subplots(1, 2, figsize=(20, 6))
axs[0].hist(zeta_rvs, density=True, bins='auto')#axs[0].hist(rvs, dencity=True, bins='auto')
axs[0].plot(x, np.exp(-0.5*x**2)*x*(x>=0), 'r')
eta=(rvs1+rvs2)
axs[1].hist(eta_rvs, density=True, bins='auto')#axs[0].hist(rvs, dencity=True, bins='auto')
axs[1].plot(x, 0*x+1/np.pi, 'r')

plt.show()
```

**–ó–∞—É–≤–∞–∂–µ–Ω–Ω—è.** –ü–µ—Ä–µ–¥–±–∞—á–∞—î—Ç—å—Å—è –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è —Ç–µ–æ—Ä–µ–º–∏ –ø—Ä–æ –æ–±—Ä–∞—Ö—É–Ω–æ–∫ —â—ñ–ª—å–Ω–æ—Å—Ç—ñ —Å—É–º–∏ –¥–≤–æ—Ö –Ω–µ–∑–∞–ª–µ–∂–Ω–∏—Ö –≤–∏–ø–∞–¥–∫–æ–≤–∏—Ö –≤–µ–ª–∏—á–∏–Ω (—á–µ—Ä–µ–∑ –∑–≥–æ—Ä—Ç–∫—É —â—ñ–ª—å–Ω–æ—Å—Ç–µ–π).

$$ùëìùúÇ(ùë•)=ùëì\xi1+\xi2(x)=\int_{-1}^{1}0.25I_{-1\leq x-t \leq 1}dt=\int_{-1}^{1}0.25I_{-1-x\leq -t \leq 1-x}dt=\int_{-1}^{1}0.25I_{-1+x\leq t \leq 1+x}(t)dt=$$
$$=I_{x\geq0}(x)\int_{-1+x}^{1}0.25dt+I_{x<0}(x)\int_{-1}^{1+x}0.25dt=I_{x\geq0}(x)(0.5-0.25x)+I_{x<0}(x)(0.5+0.25x)=0.5-0.25|x|$$


```python
# –ú—ñ—Å—Ü–µ –¥–ª—è –í–∞—à–æ–≥–æ –∫–æ–¥—É
from scipy.stats import norm, uniform
xi1=uniform(loc=-1, scale=2)
xi2=uniform(loc=-1, scale=2)
xi1_rvs=xi1.rvs(10000)
xi2_rvs=xi2.rvs(10000)

eta_rvs=xi1_rvs+xi2_rvs
x=np.arange(-2, 2+0.01, 0.01)
fig, axs=plt.subplots(1, 1, figsize=(20, 6))
axs.hist(eta_rvs, density=True, bins='auto')#axs[0].hist(rvs, dencity=True, bins='auto')
axs.plot(x, 0.5-0.25*(abs(x)), 'r')


plt.show()
```
